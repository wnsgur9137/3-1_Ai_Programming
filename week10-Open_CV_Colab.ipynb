```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
# from google.colab.patches import cv2_imshow ## colab  
img=cv2.imread('./content/test1_1920.jpg')
print('img.ndim = ', img.ndim)
print('img.shape = ', img.shape)
print('img.dtype = ', img.dtype)

# cv2_imshow(img) # colab 

#  
fig = plt.figure()
rows=1
cols=2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(img)
ax1.set_title('test1_1920.jpg')
ax1.axis('off')

img=cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
cv2.imwrite('./content/test1_half.jpg', img)
print(img.shape)

img1=cv2.imread('./content/test1_half.jpg')
gimg = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
cv2.imwrite('test1_halfgray.jpg', gimg)
print(gimg.shape)
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(gimg)
ax2.set_title('test1_halfgray.jpg')
ax2.axis('off')

plt.show()
```
